```python
# import libraries
import pandas as pd
import numpy as np
```

```python
# load datasets 
data0 =pd.read_csv("data/daily_sales_data_0.csv")
data1 =pd.read_csv("data/daily_sales_data_1.csv")
data2 =pd.read_csv("data/daily_sales_data_2.csv")

```

```python
# view data 0
data0.head(10)
```

```python
# view data 1
data1.head(10)
```

```python
# view data 2
data2.head(10)
```

```python
# Concatenating the datasets
data = [data0, data1, data2]
result = pd.concat(data)
```

```python
result.head(1100)
```

```python
# convert the price to float by removing the "$"
import re
result['price'] = [re.sub('[^0-9,.]','',x) for x in result['price']]
result['price'] = result['price'].astype('float')
```

```python
result.info()
```

```python
# create sales column
result['sales'] = result['price'] * result['quantity']
result.head()
```

```python
# get extracted data
output =result[['product','sales','date','region']]
output.to_csv("output_data.csv")
```
